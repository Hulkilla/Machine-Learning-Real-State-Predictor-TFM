```python
import pandas as pd

# Base de Datos y Variables de Entorno
from conexion import connect_to_postgresql
```

```python
engine = connect_to_postgresql()
conn = engine.raw_connection()
cur = conn.cursor()
try:
    query = """
    SELECT * FROM public.ft_indicadores;
    """
    cur.execute(query)

    df = pd.DataFrame(cur.fetchall(), columns=[desc[0] for desc in cur.description])

    conn.commit()
    print("La consulta se ha ejecutado con éxito")
except Exception as e:
    print(f"Ha ocurrido un error: {str(e)}")
finally:
    cur.close()
    conn.close()

df.head()
```

```python
df_copy = df.copy()
df_copy.head()
```

```python
df_copy.drop(columns=['Fecha', 'Habitaciones', 'Aseos', 'Precio', 'Metros', 'Comodidades'], inplace=True)
df_copy.head()
```

```python
df_copy['Municipio'] = df_copy['Municipio_Tipo'].str.split('_').str[0]
df_copy['Tipo_Vivienda'] = df_copy['Municipio_Tipo'].str.split('_').str[1]
df_copy.head()
```

```python
df_copy.drop(columns=['Municipio_Tipo'], inplace=True)
df_copy.head()
```

Por simplicidad, se va a unificar por promedio el CUDIS, Provincia, Capital y Municipio, las variables de CUDIS, Población, Precio_medio_mun_tipo y Renta bruta media por persona.

```python
grouped_df = df_copy.groupby(['CUDIS', 'Municipio', 'Provincia', 'Capital', 'Tipo_Vivienda'])[['Población', 'Precio_medio_mun_tipo', 'Renta bruta media por persona']].mean().astype(int).reset_index()

grouped_df['CUDIS'] = grouped_df['CUDIS'].astype(int)

grouped_df.head()
```

```python
grouped_df.drop_duplicates(inplace=True)
grouped_df.head()
```

```python
# Count occurrences of each municipality
municipality_counts =grouped_df['CUDIS'].value_counts()

# Filter municipalities with more than 1 occurrence
municipalities_filtered = municipality_counts[municipality_counts > 2]

municipalities_filtered
```

```python
grouped_df.rename(
    columns={
        'Población': 'Poblacion',
        'Precio_medio_mun_tipo': 'Precio_medio',
        'Renta bruta media por persona': 'Renta_bruta_media'
    },
    inplace=True)

group_mapping = {
    'Apartment': 'Piso',
    'Single-family Home': 'Casa'
}

# Crear nueva columna agrupada
grouped_df['Vivienda'] = grouped_df['Tipo_Vivienda'].map(group_mapping)
grouped_df.drop(columns=['Tipo_Vivienda'], inplace=True)
grouped_df.head()
```

```python
grouped_json = grouped_df.to_json(orient='records', force_ascii=False)

with open('indicadores.json', 'w', encoding='utf-8') as json_file:
  json_file.write(grouped_json)
```
